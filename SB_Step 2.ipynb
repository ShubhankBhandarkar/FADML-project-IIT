```python
tagged_sentence_wikdict = {}
counter = 0
for word in tagged_wikidict.keys():
    tagged_sentence_wikdict[word] = ' '.join(tagged_wikidict[word])
    counter+=1
```

```python
def create_df_from_dict(mention_page_dict, wiki_sentence_dict):
    mention_list = []
    entity_list = []
    content_list = []
    for mention in mention_page_dict.keys():
        for entity in mention_page_dict[mention]:
            try:
                content_list.append(wiki_sentence_dict[entity])
                entity_list.append(entity)
                mention_list.append(mention)
            except:
                print('Entity == ' + entity + ' == Not found')

    df = pd.DataFrame({'Mention': mention_list, 'Entity': entity_list, 'Content': content_list})
    df = df[['Mention', 'Entity', 'Content']]
    return df
```

```python
train_df = create_df_from_dict(top_30_pages_train_dict, tagged_sentence_wikdict)
train_df.head()
```
